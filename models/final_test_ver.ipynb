```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tqdm import tqdm
```

```python
df1 = df1.dropna(subset=['Order_Demand'])
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Order_Demand'] = pd.to_numeric(df1['Order_Demand'], errors='coerce')
df1 = df1.dropna(subset=['Order_Demand'])
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)]
```

```python
results = []
```

```python
for category in tqdm(df['Product_Category'].unique(), desc="Processing Categories"):
    df_cat = df1[df1['Product_Category'] == category].copy()

    df_cat = df_cat.set_index('Date').resample('W').sum().reset_index()

    df_cat = remove_outliers_iqr(df_cat, 'Order_Demand')

    if len(df_cat) < 30:
        continue

    df_prophet = df_cat.rename(columns={'Date': 'ds', 'Order_Demand': 'y'})

    train_size = int(len(df_prophet) * 0.85)
    train_df = df_prophet.iloc[:train_size]
    test_df = df_prophet.iloc[train_size:]

    model = Prophet()
    model.fit(train_df)

    future = model.make_future_dataframe(periods=len(test_df), freq='W')
    forecast = model.predict(future)

    forecast['ds'] = pd.to_datetime(forecast['ds'])
    test_df['ds'] = pd.to_datetime(test_df['ds'])
    merged = pd.merge(test_df, forecast[['ds', 'yhat']], on='ds', how='inner')
    actual = merged['y']
    pred = merged['yhat']

    mae = mean_absolute_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mape = np.mean(np.abs((actual - pred) / actual)) * 100
    r2 = r2_score(actual, pred)

    results.append({
        'Product_Category': category,
        'MAE': mae,
        'RMSE': rmse,
        'MAPE': mape,
        'R2': r2
    })
```

```python
results_df1_weekly_prophet = pd.DataFrame(results).sort_values(by='RMSE')
results_df1_weekly_prophet= results_df1_weekly_prophet.reset_index(drop=True)
results_df1_weekly_prophet
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
results_hw = []

for cat in tqdm(df1['Product_Category'].unique(), desc='Processing Categories'):
    df_cat = df1[df1['Product_Category'] == cat].copy()

    # Weekly aggregation
    df_cat = df_cat.set_index('Date').resample('W').sum().reset_index()
    df_cat = remove_outliers_iqr(df_cat, 'Order_Demand')

    if len(df_cat) < 60:
        continue  # skip small series

    df_cat = df_cat.sort_values('Date')

    # Train-test split (80-20)
    split = int(len(df_cat) * 0.8)
    train = df_cat.iloc[:split]
    test = df_cat.iloc[split:]

    try:
        model = ExponentialSmoothing(train['Order_Demand'],
                                     seasonal='add',
                                     seasonal_periods=52,  # weekly seasonality
                                     trend='add').fit()

        forecast = model.forecast(len(test))
        actual = test['Order_Demand'].values

        mae = mean_absolute_error(actual, forecast)
        rmse = np.sqrt(mean_squared_error(actual, forecast))
        mape = np.mean(np.abs((actual - forecast) / actual)) * 100
        r2 = r2_score(actual, forecast)

        results_hw.append({
            'Product_Category': cat,
            'MAE': mae,
            'RMSE': rmse,
            'MAPE': mape,
            'R2': r2
        })

    except Exception as e:
        print(f"Category {cat} failed: {e}")
        continue

```

```python
results_hw_df = pd.DataFrame(results_hw)
results_hw_df = results_hw_df.sort_values(by='RMSE')
results_hw_df = results_hw_df.reset_index(drop=True)
results_hw_df.head()
```

```python
results_hw_df.to_csv('results_hw_df.csv', index=False)
```
